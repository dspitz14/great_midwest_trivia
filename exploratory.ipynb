```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('data/merged_question.csv')
```

```python
q_counts = df['question'].groupby(df['year']).agg('count').reset_index()
```

```python
df.to_json('data/trivia_qs.json')
```

```python
type(q_counts)
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(20, 10)
sns.barplot(x="year", y="question", data=q_counts, color='blue', saturation=.15)

plt.title('Great Midwest Trivia Contest Question Count', fontsize= 20)
plt.xlabel('Year', fontsize=17)
plt.ylabel('Question Count (All Action, Off-Campus Only, etc)', fontsize=17)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

```python
df['q_length']= df['question'].map(str).apply(len)
```

```python
q_len= df['q_length'].groupby(df['year']).agg('mean').reset_index()
q_len
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(20, 10)
sns.barplot(x="year", y="q_length", data=q_len, color='blue', saturation=.15)

plt.title('Great Midwest Trivia Contest Average Question Length by Year', fontsize= 20)
plt.xlabel('Year', fontsize=17)
plt.ylabel('Average Question Length', fontsize=17)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

```python
df['comma_count'] = df.question.str.count(',')
```

```python
q_com= df['comma_count'].groupby(df['year']).agg('mean').reset_index()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(20, 10)
sns.barplot(x="year", y="comma_count", data=q_com, color='blue', saturation=.15)

plt.title('Great Midwest Trivia Contest Question Complexity', fontsize= 20)
plt.xlabel('Year', fontsize=17)
plt.ylabel('Average Number of Commas in Each Question', fontsize=17)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```
